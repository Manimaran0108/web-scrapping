```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as sk
```

```python
url = "https://www.flipkart.com/search?q=hometheater+speaker&sid=0pm%2C0o7&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&as-pos=1&as-type=RECENT&suggestionId=hometheater+speaker%7CSpeakers&requestId=48c72351-a7d9-4895-9b78-f78ccff00f20&as-searchtext=homethea"
```

```python
res=requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
hometheatre_names=soup.select(".s1Q9rs")
```

```python
hometheatre_names
```

```python
hometheatre_prices=soup.select("._8VNy32 ._30jeq3")
```

```python
hometheatre_prices
```

```python
hometheatre_offers=soup.select("._2nRPpA:nth-child(3) ._2FuKQX span , ._88-NUv+ ._2nRPpA ._2FuKQX span , ._8VNy32 span")
```

```python
hometheatre_offers
```

```python
df=sk.DataFrame(hometheatre_names,columns=['hometheatre_names'])
```

```python
df
```

```python
df['hometheatre_prices']=hometheatre_prices
```

```python
df
```

```python
df['hometheatre_offers']=hometheatre_offers
```

```python
df
```

```python
df.to_csv('skk.csv')
```
